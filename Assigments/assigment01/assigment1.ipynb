```python

rent = 25000
food = 15000
transport = 7000
utilities = 8000
total_expense = rent + food + transport + utilities

budget_limit = 60000

print("Total Monthly Expense:", total_expense)

if total_expense <= budget_limit:
    print("Status: Within Budget")
else:
    print("Status: Over Budget")
```

```python
units = int(input("Enter electricity units consumed: "))

if units <= 100:
    category = "Basic Rate"
elif units <= 300:
    category = "Standard Rate"
else:
    category = "High Usage Rate"
print("Billing Category:", category)
```

```python
attendance = [72, 85, 90, 60, 40, 78, 88]

eligible_count = 0
not_eligible_count = 0
for percentage in attendance:
    if percentage >= 75:
        eligible_count += 1
    else:
        not_eligible_count += 1
print("Number of students meeting attendance requirement (>=75%):", eligible_count)
print("Number of students NOT meeting attendance requirement (<75%):", not_eligible_count)
```

```python
cgpas = [3.6, 3.2, 3.8, 3.4, 3.9]
incomes = [75000, 85000, 60000, 90000, 70000]
eligible_count = 0
for i in range(len(cgpas)):
    if cgpas[i] >= 3.5 and incomes[i] <= 80000:
        eligible_count += 1

print("Number of students eligible for scholarship:", eligible_count)
```

```python
readings = [15, 25, 60, 85, 50, 10, 75, 90]
alert_count = 0
total_readings = len(readings)
for reading in readings:
    if reading < 20 or reading > 80:
        alert_count += 1
alert_percentage = (alert_count / total_readings) * 100

print("Total alert readings:", alert_count)
print("Percentage of alert readings:", alert_percentage, "%")
```
