```python
python_students = {"Ali", "Sara", "Ahmed", "Zain"}
ml_students = {"Sara", "Hassan", "Ali", "Fatima"}

both_courses = python_students & ml_students
print("Students enrolled in both courses:", both_courses)

only_python = python_students - ml_students
print("Students enrolled only in Python:", only_python)

master_students = python_students | ml_students
print("Master student set:", master_students)

ml_students.add("Bilal")
print("Updated ML students:", ml_students)

if "Ali" in master_students:
    print("Ali is in the master student list.")
```

```python
employees = {
    "E01": {"name": "Ali", "salary": 60000},
    "E02": {"name": "Sara", "salary": 75000},
    "E03": {"name": "Ahmed", "salary": 50000}
}

# Print all employee IDs
print("Employee IDs:")
for emp_id in employees:
    print(emp_id)

# Calculate total payroll
total_payroll = 0
for emp_id in employees:
    total_payroll += employees[emp_id]["salary"]

print("Total Payroll:", total_payroll)

# Identify employees earning above 60000
high_earners = []
for emp_id in employees:
    if employees[emp_id]["salary"] > 60000:
        high_earners.append(employees[emp_id]["name"])

print("Employees earning above 60000:", high_earners)

# Add a new employee record
employees["E04"] = {"name": "Zain", "salary": 68000}

print("Updated Employees:", employees)
```

```python
sales_data = {
    "electronics": [
        ("Laptop", 120000, 3),
        ("Mouse", 1500, 10)
    ],
    "clothing": [
        ("Shirt", 2500, 5),
        ("Shoes", 5000, 2)
    ]
}

revenue_summary = {}
unique_products = set()

for category in sales_data:
    total_revenue = 0

    for product_name, price, quantity in sales_data[category]:
        total_revenue += price * quantity
        unique_products.add(product_name)

    revenue_summary[category] = total_revenue

print("Revenue Report")
print("----------------------")

for category in revenue_summary:
    print(f"{category.capitalize()} Revenue: {revenue_summary[category]}")

print("----------------------")
print("Unique Products:", unique_products)
```
